```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("sd.csv")

# Display first rows
print(df.head())

# Engagement Rate Calculation
df['Engagement_Rate'] = (
    (df['Likes'] + df['Comment'] +
     df['Shares'] + df['Save'])
    / df['Followers']
) * 100

# Summary Statistics
print(df.describe())

# Top Content Types
sns.barplot(x='post_type',
            y='Engagement_Rate',
            data=df)
plt.title("Engagement Rate by Content Type")
plt.show()

# Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')
plt.show()
```
